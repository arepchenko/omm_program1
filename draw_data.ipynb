```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
%matplotlib inline
```

# Графики численного решения

## $ f = 2(x-x^2+y-y^2) $

## $ \alpha = 0 $

```python
with open('solution.txt') as file:
    arr = np.array([[float(digit) for digit in line[:-2].split(',')] for line in file])
```

```python
plt.figure(figsize=(11, 11))
fig = sns.heatmap(arr, cmap='jet', square=True, xticklabels=False, yticklabels=False,cbar_kws={"shrink": 0.82})

fig
```

```python
plt.figure(figsize=(10, 10))
fig = plt.imshow(arr, cmap=plt.cm.jet, extent=[0, 1,0,1])
plt.colorbar(fig,fraction=0.046, pad=0.04)
fig
```

```python
x = np.linspace(0, 1, len(arr)) 
y = np.linspace(0, 1, len(arr)) 
y, x = np.meshgrid(x, y)

    
fig = plt.figure(figsize=(10, 10))
axes = Axes3D(fig)
axes.set(xlabel=r'$x$', ylabel=r'$y$', zlabel=r'$U(x,y)$')
axes.plot_surface(x, y, arr, rstride=1, cstride=1, cmap='jet')
plt.show()
```

# Графики погрешности

## $V(x, y) = x  (1 - x)  y  (1 - y)$ - аналитическое решение

```python
with open('error.txt') as file:
    arr = np.array([[float(digit) for digit in line[:-2].split(',')] for line in file])  
```

```python
plt.figure(figsize=(11, 11))
fig = sns.heatmap(arr, cmap='jet', square=True, xticklabels=False, yticklabels=False,cbar_kws={"shrink": 0.82})

fig
```

```python
plt.figure(figsize=(10, 10))
fig = plt.imshow(arr, cmap=plt.cm.jet, extent=[0, 1,0,1])
plt.colorbar(fig,fraction=0.046, pad=0.04)
fig
```

```python
x = np.linspace(0, 1, len(arr)) 
y = np.linspace(0, 1, len(arr)) 
y, x = np.meshgrid(x, y)

fig = plt.figure(figsize=(10, 10))
axes = Axes3D(fig)
axes.set(xlabel=r'$x$', ylabel=r'$y$', zlabel=r'$\Delta U(x,y)$')
axes.plot_surface(x, y, np.abs(arr), rstride=1, cstride=1, cmap='jet')
plt.show()
```
